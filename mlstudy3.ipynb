```python
import pandas as pd
import numpy as np

df=pd.DataFrame({"name":['Kim','Lee','Park'],
                   "toy":[np.nan,'Bobby','Ironman'],
                   "born":[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]})
```

```python
df
```

```python
df.dropna(axis=1)
```

```python
df.dropna(how='all')
```

```python
df.dropna(thresh=2)
```

```python
df.dropna(subset=['name','toy'])
```

```python
df.dropna(inplace=True)
df
```
